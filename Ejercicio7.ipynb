```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0)**2 + d * (x - x0)**3


import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]

# Calculamos los coeficientes del spline para que pase por los puntos dados
s = [
    {
        "a": 1,
        "b": 7,
        "c": 0,
        "d": -3
    },
    {
        "a": 5,
        "b": -2,
        "c": 0,
        "d": 0
    },
]


for i, x_i in enumerate(xs[:-1]):
    a = s[i]["a"]
    b = s[i]["b"]
    c = s[i]["c"]
    d = s[i]["d"]
    print(
        f"S{i}(x) = {a:.2f} + {b:.2f}*(x - {x_i}) + {c:.2f}*(x - {x_i})^2 + {d:.2f}*(x - {x_i})^3"
    )

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 100)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, label=f'S{i}(x)')

plt.scatter(xs, ys, color="purple")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos que pasan por los puntos")
plt.legend()
plt.show()
```
